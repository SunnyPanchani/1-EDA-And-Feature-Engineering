```python

df3=df.copy()
df3['Purchase Amount (USD)']=df3['Purchase Amount (USD)'].fillna(df3['Purchase Amount (USD)'].mean())
df3['Review Rating']=df3['Review Rating'].fillna(df3['Review Rating'].mean())
df3['Date Purchase'] = pd.to_datetime(df3['Date Purchase'], dayfirst=True)


df3['Year'] = df3['Date Purchase'].dt.year
df3['month'] = df3['Date Purchase'].dt.month


year_counts = df3['Year'].value_counts().sort_index()

print(year_counts)
```

```python
monthly_count = df3.groupby(['Year', 'month']).value_counts()
print(monthly_count)
df3[['month','Year']].value_counts()
```

```python



year_counts = df3['Year'].value_counts().sort_index()


plt.figure(figsize=(8, 5))
year_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Number of Purchases per Year")
plt.xlabel("Year")
plt.ylabel("Purchase Count")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by Month for all years
# monthly_counts = df3['month'].value_counts().sort_index()
df3['Month'] = df3['Date Purchase'].dt.month
monthly_sales = df3.groupby(['Year', 'Month'])['Purchase Amount (USD)'].sum().unstack(level=0).fillna(0)

monthly_sales.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title("Monthly Sales by Year")
plt.xlabel("Month")
plt.ylabel("Total Sales (USD)")
plt.xticks(ticks=range(0, 12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.legend(title='Year')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Item sales

```python
item_per_year = df3.groupby(['Year', 'Item Purchased']).size().unstack(fill_value=0)
```

```python
top_items = df3['Item Purchased'].value_counts().nlargest(10).index
item_per_year = item_per_year[top_items]
```

### Top 10 Items Per Year (Grouped Bar Plot)

```python
ax = item_per_year.plot(kind='bar', figsize=(12, 6))
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=8, padding=2)


plt.title("Top 10 Item Purchases per Year")
plt.xlabel("Year")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=0)
plt.legend(title='Item', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

### Heatmap of Items per Month (Better for Large Data)

```python
import seaborn as sns


item_month = df3.groupby(['month', 'Item Purchased']).size().unstack(fill_value=0)


top_items = df3['Item Purchased'].value_counts().nlargest(10).index
item_month = item_month[top_items]


plt.figure(figsize=(12, 6))
sns.heatmap(item_month.T, cmap="YlGnBu", annot=True, fmt='d')
plt.title("Monthly Purchase Count of Top 10 Items")
plt.xlabel("Month")
plt.ylabel("Item")
plt.tight_layout()
plt.show()
```

### Item Trend Over Time (Line Plot)

```python

df3['Month_Year'] = df3['Date Purchase'].dt.to_period('M').astype(str)

item_trend = df3.groupby(['Month_Year', 'Item Purchased']).size().unstack(fill_value=0)
top_items = df3['Item Purchased'].value_counts().nlargest(5).index
item_trend = item_trend[top_items]


item_trend.plot(figsize=(14, 6))
plt.title("Monthly Trend of Top 5 Items Purchased")
plt.xlabel("Month-Year")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python



# Create folder to save images
os.makedirs("charts", exist_ok=True)

# --- 1. Bar Chart: Year-wise Purchases ---
year_counts = df3['Year'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
year_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Number of Purchases per Year")
plt.xlabel("Year")
plt.ylabel("Purchase Count")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("charts/yearly_purchases.png")
plt.close()

# --- 2. Monthly Sales by Year ---
df3['Month'] = df3['Date Purchase'].dt.month
monthly_sales = df3.groupby(['Year', 'Month'])['Purchase Amount (USD)'].sum().unstack(level=0).fillna(0)
monthly_sales.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title("Monthly Sales by Year")
plt.xlabel("Month")
plt.ylabel("Total Sales (USD)")
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.legend(title='Year')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("charts/monthly_sales_by_year.png")
plt.close()

# --- 3. Top 10 Items Per Year ---
item_per_year = df3.groupby(['Year', 'Item Purchased']).size().unstack().fillna(0)
top_10_items = df3['Item Purchased'].value_counts().nlargest(10).index
item_per_year = item_per_year[top_10_items]
ax = item_per_year.plot(kind='bar', figsize=(12, 6))
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=8, padding=2)
plt.title("Top 10 Item Purchases per Year")
plt.xlabel("Year")
plt.ylabel("Number of Purchases")
plt.legend(title='Item', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y')
plt.savefig("charts/top_10_items_per_year.png")
plt.close()

# --- 4. Heatmap: Monthly Purchase Count of Top 10 Items ---
item_month = df3.groupby(['month', 'Item Purchased']).size().unstack(fill_value=0)
top_items = df3['Item Purchased'].value_counts().nlargest(10).index
item_month = item_month[top_items]
plt.figure(figsize=(12, 6))
sns.heatmap(item_month.T, cmap="YlGnBu", annot=True, fmt='d')
plt.title("Monthly Purchase Count of Top 10 Items")
plt.xlabel("Month")
plt.ylabel("Item")
plt.tight_layout()
plt.savefig("charts/monthly_top10_heatmap.png")
plt.close()

# --- 5. Monthly Trend of Top 5 Items Purchased ---
df3['Month_Year'] = df3['Date Purchase'].dt.to_period('M').astype(str)
item_trend = df3.groupby(['Month_Year', 'Item Purchased']).size().unstack(fill_value=0)
top_items = df3['Item Purchased'].value_counts().nlargest(5).index
item_trend = item_trend[top_items]
item_trend.plot(figsize=(14, 6))
plt.title("Monthly Trend of Top 5 Items Purchased")
plt.xlabel("Month-Year")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.savefig("charts/monthly_trend_top5.png")
plt.close()

# ==========================
# EXPORT TO EXCEL
# ==========================
excel_path = "clenned_Fashion_Retail_Sales.xlsx"
with pd.ExcelWriter(excel_path, engine='xlsxwriter') as writer:
    # Optional: write data summaries
    df3.to_excel(writer, sheet_name='Fashion Sales Data', index=False)
    year_counts.to_excel(writer, sheet_name='Yearly Sales')
    monthly_sales.to_excel(writer, sheet_name='Monthly Sales')
    item_per_year.to_excel(writer, sheet_name='Top 10 Items')
    item_month.to_excel(writer, sheet_name='Heatmap Data')
    item_trend.to_excel(writer, sheet_name='Top 5 Trend')

    workbook = writer.book
    sheet = workbook.add_worksheet('Charts')
    writer.sheets['Charts'] = sheet

    # Insert images into the sheet
    chart_paths = [
        ("charts/yearly_purchases.png", "Yearly Purchases"),
        ("charts/monthly_sales_by_year.png", "Monthly Sales"),
        ("charts/top_10_items_per_year.png", "Top 10 Items"),
        ("charts/monthly_top10_heatmap.png", "Heatmap"),
        ("charts/monthly_trend_top5.png", "Top 5 Trend"),
    ]

    row = 1
    for path, label in chart_paths:
        sheet.write(row, 0, label)
        sheet.insert_image(row + 1, 0, path, {'x_scale': 0.75, 'y_scale': 0.75})
        row += 22  # Space between images
```
